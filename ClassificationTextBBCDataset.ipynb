##Transforming text using count encoding
- one method to encode text data into numeric form is to use the CountVectorizer estimator in sci-kit learn. When you use the CountVectorizer, every document in your dataset will be represented by a feature vector whose size is equal to the size of your vocabulary. 
    - word in document will be represented using a number which shows us how many times that word occurs in a document. 
    

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
x_train_count = count_vectorizer.fit_transform(data['text'])

count_vectorizer.vocabulary_

#numeric identifier gives us the position of that particular word in the feature vector representation of each document
```

```python
x_train_count.shape
#now it's (num features we have, number of unique words we have)
```

```python
print(x_train_count[15])
#looking at article at index 15 that we examined in text form earlier
#feature representation now vs. numerically encoded sparse matrix representation 
#presence of word demontrated in how often particular word occurs (you can see that on right)
```

##Transform count vectorizer into tfidf
- Calculating inverse document frequency
- another way to convert text data into numeric form
- converting data into its term frequency-inverse document frequency, that is tfidf 
- size of feature vector = size of vocab 
- each word assigned a numeric id

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
x_train_tfidf = tfidf_vectorizer.fit_transform(data['text']) 

tfidf_vectorizer.vocabulary_
```

```python
x_train_tfidf.shape
```

```python
print(x_train_tfidf[15])
#each word represented with term frequency and inverse document frequency scores
```

```python
idf_scores = list(zip(tfidf_vectorizer.get_feature_names(), tfidf_vectorizer.idf_))
```

```python
idf_scores[4000:4050]
```

##Split data into train, test data

```python
from sklearn.model_selection import train_test_split

Y = data['category']

x_train, x_test, y_train, y_test = train_test_split(x_train_tfidf, Y, test_size=0.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

##Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=10)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)

y_pred[0:10]
```

##Accuracy score

```python
from sklearn.metrics import accuracy_score

print("Accuracy : ", accuracy_score(y_test, y_pred))
```
